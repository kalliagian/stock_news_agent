## News dataset

```python
news_dataset = load_dataset("Kallia/stock-news-summaries", split="train")
news_dataset
```

```python
from datasets import DatasetDict

news_dataset = news_dataset.shuffle(seed=42)
train_valid_test = news_dataset.train_test_split(test_size=0.2, seed=42)
valid_test = train_valid_test["test"].train_test_split(test_size=0.5, seed=42)

news_dataset_dict = DatasetDict({
    "train": train_valid_test["train"],
    "validation": valid_test["train"],
    "test": valid_test["test"]
})

print(news_dataset_dict)
```

```python
from transformers import AutoTokenizer

model_checkpoint = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
inputs = tokenizer("Shares of Tesla were flat in premarket trading Thursday.")
inputs
```

```python
tokenizer.convert_ids_to_tokens(inputs.input_ids)
```

```python
max_input_length = 512
max_target_length = 50


def preprocess_function(examples):
    model_inputs = tokenizer(
        examples["article"],
        max_length=max_input_length,
        truncation=True,
    )
    labels = tokenizer(
        examples["summary"], max_length=max_target_length, truncation=True
    )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
tokenized_datasets = news_dataset_dict.map(preprocess_function, batched=True)
```

### Fine-tuning

```python
from transformers import AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
from transformers import Seq2SeqTrainingArguments

batch_size = 4 #8
num_train_epochs = 4 #8
# Show the training loss with every epoch
logging_steps = len(tokenized_datasets["train"]) // batch_size
model_name = model_checkpoint.split("/")[-1]

args = Seq2SeqTrainingArguments(
    output_dir=f"{model_name}-finetuned-stock-news",
    eval_strategy="epoch",
    learning_rate=2.5e-5, #5.6
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=2, #3
    num_train_epochs=num_train_epochs,
    predict_with_generate=True,
    logging_steps=logging_steps,
    push_to_hub=True,
    report_to="none"
)
```

```python
import numpy as np


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    # Decode generated summaries into text
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    # Replace -100 in the labels as we can't decode them
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    # Decode reference summaries into text
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    # ROUGE expects a newline after each sentence
    decoded_preds = ["\n".join(sent_tokenize(pred.strip())) for pred in decoded_preds]
    decoded_labels = ["\n".join(sent_tokenize(label.strip())) for label in decoded_labels]
    # Compute ROUGE scores
    result = rouge_score.compute(
        predictions=decoded_preds, references=decoded_labels, use_stemmer=True
    )
    # Extract the median scores
    result = {key: value * 100 for key, value in result.items()} # .mid.fmeasure
    return {k: round(v, 4) for k, v in result.items()}
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
tokenized_datasets = tokenized_datasets.remove_columns(
    news_dataset_dict["train"].column_names
)
```

```python
from transformers import Seq2SeqTrainer

trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```
